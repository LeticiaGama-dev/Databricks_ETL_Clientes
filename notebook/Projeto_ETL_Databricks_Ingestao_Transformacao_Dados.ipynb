```python
dados_clientes = [
    ("Ana Mattos de Castro", "SP", 3500, "Ativo"),
    ("Patricia da Silveira Santos","RJ", 2800, "Inativo"),
    ("Jeferson Bosi","MG", 4200,"Ativo"),
    ("Mariangela Ferreira Marques", "PR", 3900, "Ativo")
]

df_clientes = spark.createDataFrame(
    dados_clientes,
    ["nome","estado","renda mensal","status"]
)

display(df_clientes)
```

```python
from pyspark.sql.functions import col

df_transformado = df_clientes.filter( 
    col("renda mensal") > 3000
)
display(df_transformado)
```
